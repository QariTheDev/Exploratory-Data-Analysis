#### <u><b>_Definition_ </b></u>

        Exploratory Data Analysis (EDA) is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and  other data visualization methods.

#### <u><b>_History_</b> </u>

Many EDA ideas can be traced back to earlier authors, for example:


- Francis Galton emphasized order statistics and quantiles.
- Arthur Lyon Bowley used precursors of the stemplot and five-number summary (Bowley actually used a "seven-figure summary", including the extremes, deciles and quartiles, along with the median—see his Elementary Manual of Statistics (3rd edn., 1920), p. 62[11]– he defines "the maximum and minimum, median, quartiles and two deciles" as the "seven positions").
- Andrew Ehrenberg articulated a philosophy of data reduction (see his book of the same name).
- The Open University course Statistics in Society (MDST 242), took the above ideas and merged them with Gottfried Noether's work, which introduced statistical inference via coin-tossing and the median test.

#### **3. Cleaning The Data**

#### **1. Extra Columns**

##### **Dropping Columns**

We don't need unnecessary data columns like toss, highest individual wicket, Given Extras and selection so we gonna remove it.

```python
#Removing The Columns
df.drop(['Toss', 'Selection','Highest Individual wicket', 'Given Extras'], axis = 1, inplace = True)
```

##### **Joining Columns**

We will join the fours and sixes column and name it boundaries

```python
#Joining columns
df['Boundaries'] = df['Fours'] + df['Sixes']
```

##### **Shifting Columns**

We will shift the result column with POTM column. \
Also we will adjust the boundaries column

```python
#Shifting the result column with Player of the match column
result = df['Result']
df.drop('Result', axis = 1, inplace = True)
df.insert(14, 'Result', result)
```

```python
#Shifting the boundaries column with wickets lost column
boundaries = df['Boundaries']
df.drop('Boundaries', axis = 1, inplace = True)
df.insert(9, 'Boundaries', boundaries)
```

#### **2. Missing Values**

Removing the null values is an essential part during the EDA Process. It is done during this process. Removing the null values can be done using various functions and techniques.
- isnull( )
- notnull( )
- dropna( )
- fillna( )

```python
#Checking for null values
df.isnull().sum()
```

```python
#Dropping the null values of runs scored column
df.dropna(subset = ['Run Scored', 'Wicket Lost', 'Boundaries', 'Run Rate', 'Highest Score', 'Wicket Taken'], axis = 0, inplace = True)
```

We will fill the unknown Player of the Match data.

```python
#filling the unknown data
df['Player Of The Match'].fillna('Unknown', inplace = True)
```

#### **Handling Outliers**

We will handle some outliers in box plot

```python
sns.boxplot(data = df['Run Rate'])
```

```python
#Removing the outliers
df = df[df['Run Rate'] < 8]
```

```python
#boxplot
sns.boxplot(data = df['Run Rate'])
```

```python
sns.boxplot(data = df['Avg Bat Strike Rate'])
```

```python
#Using the IQR method to remove the outliers
Q1 = df['Avg Bat Strike Rate'].quantile(0.25)
Q3 = df['Avg Bat Strike Rate'].quantile(0.75)
IQR = Q3 - Q1   
df = df[(df['Avg Bat Strike Rate'] >= Q1 - 1.5*IQR) & (df['Avg Bat Strike Rate'] <= Q3 + 1.5*IQR)]
```

```python
#boxplot
sns.boxplot(data = df['Avg Bat Strike Rate'])
```

```python
sns.boxplot(data = df['Boundaries'])
```

```python
#Removing the Outliers using the IQR method
Q1 = df['Boundaries'].quantile(0.25)
Q3 = df['Boundaries'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Boundaries'] >= Q1 - 1.5*IQR) & (df['Boundaries'] <= Q3 + 1.5*IQR)]
```

```python
#boxplot
sns.boxplot(data = df['Boundaries'])
```

```python
#scatterplot
sns.scatterplot(data = df['Ground'])
```

```python
#PDF
sns.displot(data = df[(df['Ground'] == 'Lahore') | (df['Ground'] == 'Karachi') ], x = 'Run Rate', kind = 'kde', hue = 'Team')
```

```python
#CDF
sns.displot(data = df, x = 'Run Rate', kind = 'ecdf', hue = 'Team')
sns.displot(data = df, x = 'Avg Bat Strike Rate', kind = 'ecdf', hue = 'Team')
```
